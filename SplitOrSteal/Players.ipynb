### Random Player <a class="anchor" id="random"></a>

Has equal chance of splitting or stealing

```python
simulate_game(RandomPlayer(), RandomPlayer(), print_rounds=True)
```

### Nice Player <a class="anchor" id="nice"></a>

Always split

```python
simulate_game(NicePlayer(), RandomPlayer(), print_rounds=True)
```

### Greedy Player <a class="anchor" id="greedy"></a>

Always steal

```python
simulate_game(GreedyPlayer(), RandomPlayer(), print_rounds=True)
```

### Fickle Player <a class="anchor" id="fickle"></a>

Initial choice is random, then alternate between split and steal

```python
simulate_game(FicklePlayer(), RandomPlayer(), print_rounds=True)
```

### Grudge Player <a class="anchor" id="grudge"></a>

Always split. Once the opponent steals, always steal.

```python
simulate_game(GrudgePlayer(), RandomPlayer(), print_rounds=True)
```

### Copycat Player <a class="anchor" id="copycat"></a>

Initial choice is split, then copy the previous action made by the opponent

```python
simulate_game(CopycatPlayer(), RandomPlayer(), print_rounds=True)
```

### SoreLoser Player <a class="anchor" id="soreloser"></a>

Split when the player has equal to or more money than the opponent, steal otherwise

```python
simulate_game(SoreLoserPlayer(), RandomPlayer(), print_rounds=True)
```

### Cautious Player <a class="anchor" id="cautious"></a>

If the opponent has stole in the previous 2 rounds, steal. Otherwise, always split.

```python
simulate_game(CautiousPlayer(), RandomPlayer(), print_rounds=True)
```

### Copykitten Player <a class="anchor" id="copykitten"></a>

Always split at first. Plays like a Copycat Player once the opponent steals twice.

```python
simulate_game(CopykittenPlayer(), RandomPlayer(), print_rounds=True)
```

### Startegic Player <a class="anchor" id="strategic"></a>

Always start with split, steal, split, split. If the opponent has ever retaliate with steal, plays like a Copycat Player. Otherwise, always steal.

```python
simulate_game(StrategicPlayer(), RandomPlayer(), print_rounds=True)
```
