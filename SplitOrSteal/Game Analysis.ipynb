```python
import pandas as pd
from players import *
from simulator import simulate_game
```

```python
player1_list = [RandomPlayer(), NicePlayer(), GreedyPlayer(), FicklePlayer(), GrudgePlayer(), \
                CopycatPlayer(), SoreLoserPlayer(), CautiousPlayer(), CopykittenPlayer(), StrategicPlayer()]
player2_list = [RandomPlayer(), NicePlayer(), GreedyPlayer(), FicklePlayer(), GrudgePlayer(), \
                CopycatPlayer(), SoreLoserPlayer(), CautiousPlayer(), CopykittenPlayer(), StrategicPlayer()]

no_games = 100
pot = 1000
pool = 100
max_round = 10
```

```python
def get_overall_average(overall):
    average = {'win': 0, 'draw': 0, 'lose': 0, 'player money': 0, 'opponent money': 0}
    for _, result in overall.items():
        for key, val in result.items():
            average[key] += val
    for key in average:
        average[key] /= len(overall)
    return average


def get_overall(player1):
    overall = {}
    for player2 in player2_list:
        results = get_results(player1, player2, no_games)
        overall[player2.get_class()] = results
    return overall
        

def get_results(player1, player2, no_games):
    results = {'win': 0, 'draw': 0, 'lose': 0, 'player money': 0, 'opponent money': 0}
    for i in range(no_games):
        simulate_game(player1, player2, pot, pool, max_round)
        if player1.money == player2.money:
            results['draw'] += 1
        elif player1.money > player2.money:
            results['win'] += 1
        else:
            results['lose'] += 1
        results['player money'] += player1.money
        results['opponent money'] += player2.money
    results['player money'] /= no_games
    results['opponent money'] /= no_games
    return results
```

```python
all_overalls = {}

for player1 in player1_list:
    all_overalls[player1.get_class()] = get_overall(player1)
```

```python
for p1 in all_overalls:
    print(p1 + " results")
    for p2, result in all_overalls[p1].items():
        print("\t", p2, ":", result)
    print()
```

```python
for p1 in all_overalls:
    print(p1 + " average results")
    print(get_overall_average(all_overalls[p1]))
    print()
```
